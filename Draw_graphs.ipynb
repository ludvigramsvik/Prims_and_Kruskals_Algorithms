## Load a graph from Matrix

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

matrix = np.loadtxt('Graphs/test_graph.txt', dtype=int)

G = nx.from_numpy_array(matrix)

nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

```python
# Print edges
edges = G.edges()
print("Edges in the graph:", edges)
```


---
### Add random weight to the edges

```python
# add random weights to edges
for u, v in G.edges():
    G[u][v]['weight'] = np.random.randint(1, 10)

# print edges with weights
print(G.edges(data=True))
```

---
### Print the graph with the weights

```python
np.random.seed(42)
pos = nx.spring_layout(G, k=0.8, iterations=100)

nx.draw(G, pos, 
        with_labels=True, 
        node_color='lightblue', 
        edge_color='gray')

edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=7, bbox=dict(facecolor='white', edgecolor='none', alpha=1))
plt.show()
```

---
### Save the new graph to file

```python
new_matrix = nx.to_numpy_array(G)
np.savetxt('Graphs/test_graph_with_weights.txt', new_matrix, fmt='%d')
```
